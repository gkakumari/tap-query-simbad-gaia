```python
from sambad_query import *
```

```python
df = load_target_list("full_str")
source_ids = df["source_id"]
```

```python
results_df_list = []

for sid in tqdm(source_ids, desc="Querying SIMBAD"):
    full_id = f"Gaia DR3{sid}"
    
    # Check if the ID exists in SIMBAD
    if not id_exists_in_simbad(simbadtap, full_id):
        print(f"ID not found in SIMBAD: {full_id}")
        continue

    try:
        df = query_simbad_for_sources(service=simbadtap, source_ids=[sid])
        if not df.empty:
            results_df_list.append(df)
    except Exception as e:
        print(f"Query failed for {full_id}: {e}")

# Combine all non-empty results
results_df = pd.concat(results_df_list, ignore_index=True)
```

```python
results_df = reorder_columns(results_df, first_col_name="gaia_id")
grouped_df = group_by_id(results_df)
```

```python
grouped_df.to_excel("ful_star_rslts.xlsx", index=False)
grouped_df.head()
```
